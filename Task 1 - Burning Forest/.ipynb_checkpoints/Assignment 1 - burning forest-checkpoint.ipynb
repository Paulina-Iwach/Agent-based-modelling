# Forest fire propagation model 

```python
class BurningForest:
    
    
     def __init__(self, L, p, start_edge):
        """
        Initializes the BurningForest symulation.

        :param L: The size of the grid (LxL)
        :param p: The probability of a tree existing in each cell.
        :param start_edge: The edge from which the fire starts. It can be "up", "down", "left", or "right".
        """
        self.L = L
        self.p = p
        self.start_edge = start_edge
        self.grid = self.CreateGrid() # Create the initial grid with trees and empty spaces.
        self.stages=[] # Stores each stage of the simulation for animation.

    
     def CreateGrid(self):
        """
        Creates a grid representing the forest where each cell is either a tree (1) or empty (0).
        The probability p determines the likelihood of a tree being in a given cell.
        
        :return: 2D numpy array representing the forest before burning.
        """
        self.grid = np.random.random((self.L,self.L))
        self.grid[self.grid >= self.p] = 0
        self.grid[self.grid > 0] = 1
        return self.grid
    
    
     def InitBurnEdge(self):
        """
        Initiates the fire at the specified edge (up, down, left, or right) of the grid.
        The fire is represented by the value 2 in the grid.
        """
        if self.start_edge == "up":
            for j in range(self.L):
                if self.grid[0, j] == 1: 
                    self.grid[0, j] = 2
        elif self.start_edge == "down":
            for j in range(self.L):
                if self.grid[self.L-1, j] == 1:  
                    self.grid[self.L-1, j] = 2
        elif self.start_edge == "left":
            for j in range(self.L):
                if self.grid[j, 0] == 1:  
                    self.grid[j, 0] = 2
        elif self.start_edge == "right":
            for j in range(self.L):
                if self.grid[j, self.L-1] == 1:  
                    self.grid[j, self.L-1] = 2
                    
    
     def Get_neighbours(self,i,j,d=1):
        """
        Retrieves the neighboring cells within a specified distance d.

        :param i: Row index of the current cell.
        :param j: Column index of the current cell.
        :param d: Distance to search for neighbors (value 1-immediate neighbors).

        :return: A list of tuples that contains the value and coordinates of a neighboring cell.
        """   
            
        neighbours = []
        for x in range(max(0, i - d), min(self.grid.shape[0], i + d + 1)): #Ensures that the row index does not go below 0
            for y in range(max(0, j - d), min(self.grid.shape[1], j + d + 1)): #and not exceed the grid size.
                if (x, y) != (i, j):
                    neighbours.append((self.grid[x, y], (x, y)))   
        return neighbours
    
    
     def MoveFire(self):
        """
        Simulates one step of fire spreading in the forest.
        """
        burning_cells = np.where(self.grid == 2) # Find all cells that are currently burning (value 2)
        self.grid[burning_cells]=3
        for i, j in zip(burning_cells[0],burning_cells[1]):
            for value, neighbour in self.Get_neighbours(i, j):
                if value == 1: # If the neighboring cell is a tree, it catches fire.
                    self.grid[neighbour[0],neighbour[1]]=2
                    
        self.stages.append(np.copy(self.grid)) # Save grid state for animation purposes.
        
        #for i in range(self.L):
         #   for j in range(self.L):
            
                #if self.grid[i, j] == 2: # Burning cells become empty
                    #burning_cells.append([i,j])
                    #grid_after_fire[i, j] = 3
            
                #elif self.grid[i, j] == 1 and self.Burning_neighbor(i, j):
                 #   grid_after_fire[i, j] = 2  # Tree catches fire if a neighbor is burning
        
        #self.grid = grid_after_fire 
        #return grid_after_fire 
    
    
     def SimulateFire(self):
        """
        Runs the fire simulation until the fire can no longer spread. 
        The simulation stops when no changes occur between consecutive steps.

        :return: Final 2D numpy array grid after the fire has burned out.
        """
        self.stages.append(np.copy(self.grid))
           
        previous_grid = self.grid.copy()
        
        while True:
            self.MoveFire()
            if np.array_equal(self.grid, previous_grid):
                break  
            previous_grid = self.grid.copy()
            
        return self.grid

        
       
     def AnimateFire(self):
        """
        Animates the fire spreading over time using matplotlib.animation.FuncAnimation.
        """
        
        if not self.stages:
            print("No stages recorded. Please run SimulateFire first.")
            return  # Exit if there are no stages to animate

        fig, ax = plt.subplots(figsize=(8, 8))
        cmap = mcolors.ListedColormap(['beige', 'forestgreen', 'firebrick', 'black'])
        bounds = [0, 1, 2, 3, 4]
        norm = mcolors.BoundaryNorm(bounds, cmap.N)

        img = ax.imshow(self.stages[0], cmap=cmap, norm=norm)

        def UpdateGrid(frame):
            img.set_array(self.stages[frame])
            return [img]

        animation = FuncAnimation(fig, UpdateGrid, frames=len(self.stages), interval=500, blit=True)
        animation.save(f"animation_L{str(self.L)}_p{str(self.p)}.gif", writer=PillowWriter(fps=5))

        plt.title(f"Simulation of Fire Spread in a Forest\nL={self.L}, p={self.p}")
        cbar = plt.colorbar(img, ticks=[0.5, 1.5, 2.5, 3.5], label='Cell States')  # Create the colorbar and assign it to `cbar`
        cbar.set_ticklabels(['0: Empty', '1: Tree', '2: Fire', '3: Burnt'])
        plt.show()
        
        
     def IfHitEdge(self):
        """
        Checks if the fire has reached the opposite edge from where it started.

        :return: True if the fire reached the opposite edge, otherwise False.
        """
       
        if not np.any(self.grid == 3):   # If no burned cells exist, return False
            return False
    
        if self.start_edge == "up" and any(self.grid[self.L-1, j] == 3 for j in range(self.L)):
            return True
        elif self.start_edge == "down" and any(self.grid[0, j] == 3 for j in range(self.L)):
            return True
        elif self.start_edge == "left" and any(self.grid[i, self.L-1] == 3 for i in range(self.L)):
            return True
        elif self.start_edge == "right" and any(self.grid[i, 0] == 3 for i in range(self.L)):
            return True
        else:
            return False
      
     
        

            
#print(forest.grid)
#forest.InitBurnEdge()
#print(forest.grid)
#print(forest.SimulateFire())
#print(forest.stages)
forest = BurningForest(L=20, p=0.2, start_edge="left")
forest.InitBurnEdge()
forest.SimulateFire()
forest.AnimateFire() 
```

#  Finding the critical value p*

```python
def FindTreshold(L, edge):
    p_list = np.arange(0, 1, 0.01)
    results = []
    
    cluster_mean =[]
    
    for p in p_list:
        hit_count = 0
        cluster_size =[] 
        for _ in range(100):
            forest = BurningForest(L=L, p=p, start_edge=edge)
            forest.InitBurnEdge() 
            forest.SimulateFire()
            cluster_size.append(largest_cluster_size(forest.grid))
            if forest.IfHitEdge(): 
                hit_count += 1
        
        results.append((p, hit_count))
        
        cluster_mean.append((p,np.mean(cluster_size)))
    return results, cluster_mean
        
L20,cluster_mean20  = FindTreshold(20,"left")
L50, cluster_mean50 = FindTreshold(50,"left")
L100, cluster_mean100 = FindTreshold(100,"left")
```

```python
def save_results(filename, results):
    df = pd.DataFrame(results, columns=['Probability', 'number'])
    df.to_csv(filename, index=False)

save_results('results_L20.csv', L20)
save_results('results_L50.csv', L50)
save_results('results_L100.csv', L100)
save_results('clusters_L20.csv', cluster_mean20)
save_results('clusters_L50.csv', cluster_mean50)
save_results('clusters_L100.csv', cluster_mean100)
```

```python
def Findp(x_val,y_val):
    result=[]
    for x,y in zip(x_val,y_val):
        if y>=0.9:
            result.append([x,y])
    if result:
        min_x, min_y = min(result, key=lambda pair: pair[0]) 
        return min_x, min_y
    else:
        return None
```

```python
p20, _ = Findp(x20,y20) 
p50, _ = Findp(x50,y50) 
p100, _ = Findp(x100,y100) 
```

```python
plt.figure(figsize=(10, 8))

plt.plot(x20, y20,marker='h', label=f'L=20 (p*={p20})', color='darkturquoise')
plt.plot(x50, y50, marker='*',label=f'L=50 (p*={p50})', color='darkorange')
plt.plot(x100, y100, marker='^',label=f'L=100 (p*={p100})',color='green')
plt.legend()
plt.xlabel('Probability of tree occurrence (p)', fontsize=14)
plt.ylabel('Probability of hitting the opposite edge', fontsize=14)
plt.grid()
```

# Plot size of the biggest cluster of burned trees as a function of p

```python
import numpy as np

def largest_cluster_size(grid):
    """
    Find the size of the largest cluster of burnt trees (represented by 3) in a grid.
    :param grid: 2D numpy array representing the forest (0 - empty, 1 - tree, 3 - burnt)
    :return: The size of the largest cluster.
    """
    L = grid.shape[0]
    labels = np.zeros_like(grid, dtype=int) # Label array of the same size as the grid
    current_label = 0  # Largest used label
    union_find = {}  # Dictionary for union-find data structure key:'new' cluster, value: number of final cluster 

    def assign_label(i, j):
        nonlocal current_label
        neighbours = get_neighbor_labels(i, j)
        if not neighbours:  # If no neighbors, create a new label
            current_label += 1
            labels[i, j] = current_label
            union_find[current_label] = current_label
        else:
            min_label = min(neighbours) # Get the minimum label from neighbors
            labels[i, j] = min_label
    
            for label in neighbours: # Perform unions for all neighbor labels
                if label != min_label:
                    union(label, min_label)

    def get_neighbor_labels(i, j):
        neighbours = []
        for x in range(max(0, i - 1), min(grid.shape[0], i + 1 + 1)):
            for y in range(max(0, j - 1), min(grid.shape[1], j + 1 + 1)):
                if (x, y) != (i, j) and labels[x,y]>0:
                    neighbours.append(labels[x, y])
                    #print(neighbours)
        return neighbours


    def union(x, y):
        #If two different label union to label y
        root_x = find(x)
        root_y = find(y)
        if root_x != root_y:
            union_find[root_x] = root_y

    def find(x):

        root = x #cluster number
        while union_find[root] != root:
            root = union_find[root]
        while x != root:
            parent = union_find[x]
            union_find[x] = root
            x = parent
        return root

    # Step 1: Assign labels to burnt trees (3)
    points_to_analyze = np.where(grid == 3)
    for i, j in zip(points_to_analyze[0], points_to_analyze[1]):
        assign_label(i, j)

    # Step 2: Resolve unions
    points_to_analyze = np.where(labels > 0)
    for i, j in zip(points_to_analyze[0], points_to_analyze[1]):
        labels[i, j] = find(labels[i, j])

    # Step 3: Count sizes of clusters
    unique, counts = np.unique(labels[labels > 0], return_counts=True)
    if len(counts) == 0:
        return 0
    
    
    #print(union_find)
    return max(counts)


forest_grid = np.array([
    [0, 0, 1, 0],
    [3, 3, 0, 0],
    [1, 3, 3, 1],
    [0, 0, 0, 3]
])



forest = BurningForest(L=10, p=0.3, start_edge="left")
forest.InitBurnEdge()
forest.SimulateFire()
forest.AnimateFire() 
print(largest_cluster_size(forest.grid)) 
```

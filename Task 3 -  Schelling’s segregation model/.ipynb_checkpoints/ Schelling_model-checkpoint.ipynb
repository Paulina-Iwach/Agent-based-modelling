```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random
from matplotlib.animation import PillowWriter
import matplotlib.animation as animation
import matplotlib.colors as mcolors
```

```python

class SchellingModel:
    def __init__(self, grid_size=100, N=10, jr=0.5, jb=0.5,m=1, record_stages=True): #N=Nb=Nr agents from one group jt=treshold
        """
        Initializes the Schelling Segregation Model.
        
        :param grid_size: Size of the square grid (default 100x100).
        :param N: Number of agents per group (total agents = 2N).
        :param jr: Satisfaction threshold for red agents, between 0 and 1.
        :param jb: Satisfaction threshold for blue agents, between 0 and 1.
        :param m: Neighborhood radius, in which agent checks out their neighbours.
        :param record_stages: Boolean flag to store snapshots at each step if True (for visualization).
        """
        self.grid_size = grid_size
        self.N = N
        self.grid = None
        self.agents = None
        self.jr = jr
        self.jb = jb
        self.m=m
        self.stages = []
        self.record_stages=record_stages


    def initialize_grid(self):
        """Initializes the grid by placing agents of two types (1-red and 2-blue) randomly on the grid."""
        
        self.grid = np.zeros((self.grid_size, self.grid_size), dtype=int)
        positions = np.random.choice(self.grid_size**2, 2 * self.N, replace=False)
        positions = [(pos // self.grid_size, pos % self.grid_size) for pos in positions]

        types = [1] * self.N + [2] * self.N
        np.random.shuffle(types)
        self.agents = list(zip(positions, types))

        for (x, y), agent_type in self.agents:
            self.grid[x, y] = agent_type
            
            
    def get_neighbours(self, x, y,m):
        """
        Finds the neighbourhood of a cell at a point (x, y) within a radius m using periodic boundary conditions.
        
        :param x: X-coordinate of the agent.
        :param y: Y-coordinate of the agent.
        :param m: Radius defining the neighborhood size.
        :return: List with the neighbour values of the selected cell and with agent's cell set to 0.
        """
        
        x_indices = (np.arange(x - m, x + m + 1) % self.grid_size)
        y_indices = (np.arange(y - m, y + m + 1) % self.grid_size)

        neighborhood = self.grid[np.ix_(x_indices, y_indices)]
        neighborhood[m, m] = 0
        
        return neighborhood.flatten()

    
    def find_empty_cell(self):
        """
        Finds a random empty cell (equal 0) on the grid.
        :return: Tuple with coordinates of an empty cell or None if no empty cells are available.
        """
        
        #empty_cells = list(zip(*np.where(self.grid == 0)))  # Locate empty cells
        #return random.choice(empty_cells) if empty_cells else None
        
        return tuple(random.choice(np.argwhere(self.grid == 0)))


    def calculate_SNI(self, x, y, agent_type):        
        """
        Calculates the Similar Neighbor Index (SNI) for an agent, based on the proportion of neighbors of the same type.
        
        :param x: X-coordinate of the agent.
        :param y: Y-coordinate of the agent.
        :param agent_type: Type of the agent (1 for red, 2 for blue).
        :return: Float representing the proportion of neighbors with the same type, 0 if no valid neighbors.
        """
        
        #neighbours = self.get_neighbours(x, y, self.m)
        #same_type_count = sum(1 for val in neighbours if val == agent_type)
        #valid_neighbours = sum(1 for val in neighbours if val != 0)
        #return same_type_count / valid_neighbours if valid_neighbours else 0
        neighbours = self.get_neighbours(x, y, self.m)
        neighbours = {1: np.count_nonzero(neighbours == 1),
                     2: np.count_nonzero(neighbours == 2)}
        
        return neighbours[agent_type]/(neighbours[1]+neighbours[2]) if neighbours[agent_type] else 0
    
    
    
    def calculate_SI(self):
        """
        Computes the Segregation Index (SI) as the average Similar Neighbor Index across all agents.
        :return: Float representing the overall segregation index.
        """
        
        sni_values = [self.calculate_SNI(x,y,agent_type) for (x, y), agent_type in self.agents ]
        SI = sum(sni_values) / (2 * self.N)
        return SI
    

    def check_relocation_condtion(self, agent_type, SNI_value):
        """
        Determines whether an agent satisfies a condition (should stay in the current position) based on colour and SNI.
        
        :param agent_type: Type of the agent (1 for red, 2 for blue).
        :param SNI_value: Similar Neighbor Index for the agent.
        :return: True if the agent is unsatisfied and should move; False otherwise.
        """
        
        if agent_type == 1: 
            return SNI_value < self.jr  
        elif agent_type == 2: 
            return SNI_value < self.jb  
        return False
        
        
    def move_agent(self, x, y, agent_type):
        """
        Relocates an unsatisfied agent to a random empty cell if available.
        
        :param x: Current X-coordinate of the agent.
        :param y: Current Y-coordinate of the agent.
        :param agent_type: Type of the agent (1 for red, 2 for blue).
        :return: Tuple with the new position of the agent or its original position if no empty cell is found.
        """
        
        new_pos = self.find_empty_cell()
        if new_pos:
            new_x, new_y = new_pos
            self.grid[x, y], self.grid[new_x, new_y] = 0, agent_type
            return (new_x, new_y), agent_type
        return (x, y), agent_type


    def update_positions(self):
        """
        Updates positions of all agents in a single simulation step.
        :return: True if any agents moved, False if no moves were made.
        """
        
        previous_agents = self.agents.copy()
        change = False
        for idx, ((x, y), agent_type) in enumerate(previous_agents):
            SNI = self.calculate_SNI(x, y, agent_type)
            if self.check_relocation_condtion(agent_type, SNI):#_values[idx]):
                self.agents[idx]= self.move_agent(x, y, agent_type)
                change=True
             
        if self.record_stages:
            self.stages.append(self.grid.copy())
    
        return change
        
    
    def run_simulation(self, max_iterations=500):
        """
        Runs the model until agents are satisfied or a maximum number of steps is reached.
        
        :param max_iterations: Maximum number of steps to simulate.
        :return: Number of cycles completed before all agents were satisfied or limit was reached.
        """
        cycle_count = 0
        while cycle_count < max_iterations:
            if not self.update_positions():
                break
            cycle_count += 1

        return cycle_count


    def display_grid(self):
        """Visualizes the current state of the grid using a color map."""
    
        cmap = mcolors.ListedColormap(['beige', 'firebrick', 'royalblue']) 
        bounds = [0, 1, 2, 3] 
        norm = mcolors.BoundaryNorm(bounds, cmap.N)

        plt.imshow(self.grid, cmap=cmap, norm=norm)
        plt.title("Schelling Model Simulation (Red & Blue)")
        plt.colorbar(ticks=[0, 1, 2], label='Cell States')
        plt.show()


    def animate(self, save_to_file=None):
  
        if not self.stages:
            print("No stages recorded. Please run the simulation steps first.")
            return
        
        fig, ax = plt.subplots(figsize=(6, 6))
        cmap = mcolors.ListedColormap(['beige','firebrick', 'royalblue'])
        num_rows, num_cols = self.grid.shape

        img = ax.imshow(self.stages[0], cmap=cmap, interpolation='none')
        ax.set_xticks([]), ax.set_yticks([])

        def update_frame(frame):
            img.set_array(self.stages[frame])
            ax.set_title(f"Schellingâ€™s segregation model\nN={self.N}, jb={self.jb}, jr={self.jr}, m={self.m}, T={frame}", 
                     fontsize=14)
            return [img]

        ani = animation.FuncAnimation(fig, update_frame, frames=len(self.stages), interval=500, blit=True)

        if save_to_file:
            ani.save(save_to_file, writer=PillowWriter(fps=5))
            print(f"Animation saved to file: {save_to_file}")
        else:
            plt.show()
```

```python
model1 = SchellingModel(grid_size=100, N=250, jr=0.5, jb=0.5, m=1)
model1.initialize_grid()
model1.run_simulation(max_iterations=500)
SI = model1.calculate_SI()
print(f"Final segregation index: {SI:.2f}")
model1.animate(save_to_file="schelling_model_ani1.gif")
```

# Part 2

```python
from joblib import Parallel, delayed
```

```python
def run_single_simulation_T(N, jr, jb, m=1):
    model = SchellingModel(N=N, jr=jr, jb=jb, m=m, record_stages=False)
    model.initialize_grid()
    cycle_count = model.run_simulation(max_iterations=100)
    return cycle_count  

def run_simulations_T(num_runs=100, n_jobs=-1):
    population_range = range(250, 4001, 250)
    results = {}
    for N in population_range:
        results_list = Parallel(n_jobs=n_jobs)(delayed(run_single_simulation_T)(N, 0.5, 0.5, 1) for _ in range(num_runs))

        results[N] = np.mean(results_list)

    return results

def save_results_to_file(results, filename="simulation_results_T(N).txt"):
    with open(filename, "w") as file:
        file.write("Population Size (N), Average Stabilization Cycles (T)\n")
        for N, T in results.items():
            file.write(f"{N}, {T}\n")
    print(f"Results have been saved to {filename}")
```

```python
def read_results_from_file(filename):
    """Reads simulation results from a .txt file and returns them as a dictionary."""
    results = {}
    with open(filename, "r") as file:
        next(file) 
        for line in file:
            N, T = line.strip().split(", ")
            results[int(N)] = float(T)

    return results
```

```python
results = run_simulations_T(num_runs=100, n_jobs=-1)
save_results_to_file(results, filename="simulation_results_T(N).txt")
data = read_results_from_file(filename="simulation_results_T(N).txt")
```

```python
def plot_T_vs_N(results):
    """Plots T (average number of cycles) against population size N."""
    populations = list(results.keys())
    average_cycles = list(results.values())

    plt.figure(figsize=(8, 5))
    plt.plot(populations, average_cycles, marker='o', color='hotpink', linestyle='-')
    plt.xlabel('N')
    plt.ylabel('T[MCS]')
    plt.title('Stabilization time (T) depending on population size (N) for M=100')
    plt.grid(True)
    plt.show()
    
plot_T_vs_N(data)
```

# SI(j)

```python
def run_single_simulation_SI(N, jr, jb, m=1):
    model = SchellingModel(N=N, jr=jr, jb=jb, m=m, record_stages=False)
    model.initialize_grid()
    model.run_simulation(max_iterations=10000)
    SI = model.calculate_SI()
    return SI

def run_simulations_SI_j(num_runs=100, n_jobs=-1):
    jt_values = np.arange(0.1, 0.81, 0.1)  # Threshold values from 0.1 to 0.9
    results = {}
    for N in [1000, 3000]:
        results_N = {}
        for j in jt_values:
            results_list = Parallel(n_jobs=n_jobs)(
                delayed(run_single_simulation_SI)(N, j, j, 1) for _ in range(num_runs)
            )
            results_N[j] = np.mean(results_list)  # Average SI value for this `j`
        
        results[N] = results_N
    
    return results

```

```python
def save_results_to_file_SI(results, filename="simulation_results_SI(j).txt"):
    with open(filename, "w") as file:
        file.write("N, j, SI\n")
        for N, j_results in results.items():
            for j, SI in j_results.items():
                file.write(f"{N}, {j}, {SI}\n")
    print(f"Results have been saved to {filename}")

def read_results_from_file_SI(filename="simulation_results_SI(j).txt"):
    results = {}
    with open(filename, "r") as file:
        next(file)  # Skip the header
        for line in file:
            N, j, SI = line.strip().split(", ")
            N, j, SI = int(N), float(j), float(SI)
            
            if N not in results:
                results[N] = {}
            results[N][j] = SI

    return results

```

```python
results_Si_j = run_simulations_SI_j(num_runs=100, n_jobs=-1)
save_results_to_file(results_Si_j, filename="results_SI(j).txt")
#data_SI_j = read_results_from_file(filename="results_SI(j).txt")
```

```python
save_results_to_file_SI(results_Si_j, filename="results_SI(j).txt")
```

```python
plt.figure(figsize=(10, 6))
colors = {1000: 'steelblue', 3000: 'orchid'}

for N in [1000,3000]:
    if N in results_Si_j:
        j_values = sorted(results_Si_j[N].keys())  # Sorting j values
        SI_values = [results_Si_j[N][j] for j in j_values]
            
        plt.plot(j_values, SI_values, marker='o', label=f'N = {N}',color=colors[N])
    
plt.xlabel('j')
plt.ylabel('SI')
plt.title('Segregation Index depending on threshold (j)')
plt.legend(title='Population Size')
plt.grid(True)
plt.show()
```

# SI(m)

```python
def run_single_simulation_SI(N, jr, jb, m=1):
    model = SchellingModel(N=N, jr=jr, jb=jb, m=m, record_stages=False)
    model.initialize_grid()
    model.run_simulation(max_iterations=100)
    SI = model.calculate_SI()
    return SI

def run_simulations_SI_m(num_runs=100, n_jobs=-1):
    mt_values = range(1,6)  
    results = {}
    for N in [1000, 3000]:
        results_N = {}
        for m in mt_values:
            results_list = Parallel(n_jobs=n_jobs)(
                delayed(run_single_simulation_SI)(N, 0.5, 0.5, m) for _ in range(num_runs)
            )
            results_N[m] = np.mean(results_list)  # Average SI value for this `j`
        
        results[N] = results_N
    
    return results
```

```python
def save_results_to_file_SIm(results, filename="simulation_results_SI(m).txt"):
    with open(filename, "w") as file:
        file.write("N, m, SI\n")
        for N, m_results in results.items():
            for m, SI in m_results.items():
                file.write(f"{N}, {m}, {SI}\n")
    print(f"Results have been saved to {filename}")

def load_results_from_file(filename = "results_SI(m).txt"):
    results_Si_m = {}
    with open(filename, 'r') as file:
        lines = file.readlines()[1:]  # Skip the header line
        for line in lines:
            N_str, values_str = line.strip().split(', ', 1)
            N = int(N_str)
            values_dict = eval(values_str)  # Convert string dict to actual dictionary
            results_Si_m[N] = values_dict
    return results_Si_m
```

```python
results_Si_m = run_simulations_SI_m(num_runs=100, n_jobs=-1)
save_results_to_file(results_Si_m, filename="results_SI(m).txt")
results_m= load_results_from_file(filename="results_SI(m).txt")
```

```python
plt.figure(figsize=(10, 6))
colors = {1000: 'steelblue', 3000: 'orchid'}

for N in [1000,3000]:
    if N in results_m:
        m_values = sorted(results_Si_m[N].keys())  # Sorting j values
        SI_values = [results_Si_m[N][m] for m in m_values]
            
        plt.plot(m_values, SI_values, marker='o', label=f'N = {N}',color=colors[N])
    
plt.xlabel('m')
plt.ylabel('SI')
plt.title('Segregation index depending on the neighbourhood radius (m) ')
plt.legend(title='Population Size (N)')
plt.grid(True)
plt.show()
```
